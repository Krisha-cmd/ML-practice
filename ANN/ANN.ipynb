# ANN classification and regression

ANN typically stands for Artificial Neural Network, which is a computational model inspired by biological neural networks. ANNs are used in machine learning and are made up of interconnected nodes (neurons) that process information and learn patterns from data.

MLP generally refers to a Multi-Layer Perceptron, which is a type of artificial neural network consisting of multiple layers of nodes (neurons). In an MLP, each node in one layer is connected to every node in the next layer, and it includes an input layer, one or more hidden layers, and an output layer.

In MLP between input and output (both having different dimensions) there can be more than one non-linear layers called hidden layers.

Weighted_linear_summation(input_features)->hidden_layer->nonlinear_activation_function->output_layer

Advantages
- non-linear
- can learn in real time by line by line training or partial fit

Disadvantages
- tuning of hyper parameters
- feature scaling sensitivity

```python
import pandas as pd
```

```python
df=pd.read_csv('../datasets/processed')
```

```python
df.head()
```

```python
X=df.drop('income',axis=1)
y=df['income']

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.3,random_state=42)
```

From sklearn docs lets perform

MLP Classification, MLPClassifier.coefs_ shows the matrix of hidden layers and the weights of each layer.
The supported loss function is corss-entropy function which is based on probability estimates.


```python
from sklearn.neural_network import MLPClassifier 
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(5, 2), random_state=1)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.coefs_
```

```python
[coef.shape for coef in clf.coefs_]
```

We have one input layer one hidden layer and one output layer.

MLPClassifier supports multiclass classification using softmax activation function
